### Reading the GFALL Kurucz file ###

```python
from carsus.io.kurucz import GFALLReader
```

Download gfall.dat from http://kurucz.harvard.edu/linelists/gfall/

```python
gfall_reader = GFALLReader('gfall.dat')
```

```python
lines_lower_idx = gfall_reader.lines.set_index(['atomic_number', 'ion_charge', 'energy_lower', 'j_lower'])
```

```python
levels_unique_idxed = gfall_reader.levels.reset_index().set_index(['atomic_number', 'ion_charge', 'energy', 'j'])
```

```python
levels_unique_idxed
```

```python
lines_lower_idx['level_index_lower'] = levels_unique_idxed['level_index']
```

```python
lines_upper_idx = lines_lower_idx.reset_index().set_index(['atomic_number', 'ion_charge', 'energy_upper', 'j_upper'])
```

```python
lines_upper_idx['level_index_upper'] = levels_unique_idxed['level_index']
```

```python
lines = lines_upper_idx.reset_index().set_index(['atomic_number', 'ion_charge', 'level_index_lower', 'level_index_upper'])
```
